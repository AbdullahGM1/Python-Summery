Slicing String

```python
word="banana!!"

print(word[0:3]) #The second number is up to but not include, up to index (2)
print(word[0:1]) #The second number is up to but not include, this is way it prints only (b)
print(word[3:9])
print(word[0:112]) #If the second number is beyond the end of the string, it stops at the end
```

```python
#If we leave off the first number or the last number of the slice, it is assumed to be the beginning or end of the string respectively

print(word[:4])
print(word[2:])
print(word[:])
```
